# Random Forest Regressor

```python
#importing necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

importing the dataset

```python
pollution=pd.read_csv("C:/Users/ELCOT/Downloads/CSV Dataset/pollution_india_2010.csv")
pollution.head()
```

Analysing and preprocessing the dataset

```python
pollution.info()
```

```python
new_data=pollution.replace({'Null':0})
new_data.sample(10)
```

```python
a=pd.to_numeric(new_data.NO2)
b=pd.to_numeric(new_data.SO2)
feature=pd.DataFrame([a,b])
x=feature.T
x.head()
```

```python
y=pd.to_numeric(new_data.PM10)
y.head()
```

Splitting the data and training the model

```python
from sklearn.model_selection import train_test_split, GridSearchCV
x_train, x_test, y_train, y_test=train_test_split(x,y,test_size=0.2, random_state=42)
params={'criterion':['squared_error','friedman_msc','poission']}
```

```python
from sklearn.ensemble import RandomForestRegressor
model=RandomForestRegressor()
grid_model=GridSearchCV(model,params,scoring='neg_mean_squared_error',cv=5)
grid_model.fit(x_train,y_train)
```

```python
# model prediction
y_pred=grid_model.predict(x_train)
y_pred
```

```python
#accuracy of the model using r2_score
from sklearn.metrics import r2_score
print(r2_score(y_train,y_pred)*100)
```

### using random forest Regressor achived 83% accuracy
